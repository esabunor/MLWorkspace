```python
def dialogue_act_features(posts, i):
    featuresets = {}
    post = posts[i]
    tokens = nltk.word_tokenize(post.text)
    freq = nltk.FreqDist([ tag for (word, tag) in nltk.pos_tag(tokens)])
    featuresets.update({key: freq[key] for key in freq.keys()})
    for word in tokens:
        featuresets['contains(%s)' % word.lower()] = True
    return [featuresets, post.get('class')]
        
```

```python
def rte_features(rtepair):
    extractor = nltk.RTEFeatureExtractor(rtepair)
    features = {}
    features['word_overlap'] = len(extractor.overlap('word'))
    features['word_hyp_extra'] = len(extractor.hyp_extra('word'))
    features['ne_overlap'] = len(extractor.overlap('ne'))
    features['ne_hyp_extra'] = len(extractor.hyp_extra('ne'))
    return features
```

```python
import math
def entropy(labels):
 freqdist = nltk.FreqDist(labels)
 probs = [freqdist.freq(l) for l in nltk.FreqDist(labels)]
 return sum([p * math.log(p,2) for p in probs])
```

```python
def find_chunks(string, tagged_sents):
    cp = nltk.RegexpParser('CHUNK: %s'% string)
    for sent in tagged_sents:
        tree = cp.parse(sent)
        for subtree in tree.subtrees():
            if subtree.label() == 'CHUNK':
                print(tree)

find_chunks('{<V.*> <TO> <V.*>}', nltk.corpus.brown.tagged_sents())
```

<h4>Using UnigramTagger for Tagging chunks</h4>

```python
class ConsecutiveNPChunkTagger(nltk.TaggerI):
     def __init__(self, train_sents):
         train_set = []
         for tagged_sent in train_sents:
             untagged_sent = nltk.tag.untag(tagged_sent)
             history = []
             for i, (word, tag) in enumerate(tagged_sent):
                 featureset = npchunk_features(untagged_sent, i, history)
                 train_set.append( (featureset, tag) )
                 history.append(tag)
         self.classifier = nltk.MaxentClassifier.train(train_set, algorithm='megam', trace=0)
        
     def tag(self, sentence):
         history = []
         for i, word in enumerate(sentence):
             featureset = npchunk_features(sentence, i, history)
             tag = self.classifier.classify(featureset)
             history.append(tag)
         return zip(sentence, history)
    
    
class ConsecutiveNPChunker(nltk.ChunkParserI):
     def __init__(self, train_sents):
         tagged_sents = [[((w,t),c) for (w,t,c) in
         nltk.chunk.tree2conlltags(sent)]
         for sent in train_sents]
         self.tagger = ConsecutiveNPChunkTagger(tagged_sents)
            
     def parse(self, sentence):
         tagged_sents = self.tagger.tag(sentence)
         conlltags = [(w,t,c) for ((w,t),c) in tagged_sents]
         return nltk.chunk.conlltags2tree(conlltags)
```

```python
query = 'What cities are in China'.split()
```

```python
trees = cp.parse(query)
def getNodes(trees):
    for node in trees:
        if type(node) is nltk.Tree:
            if node.label() == 'ROOT':
                print("======== Sentence =========")
                print("Sentence:", " ".join(node.leaves()))
            else:
                print("Label:", node.label()['SEM'])
                print("Leaves:", node.leaves())

            getNodes(node)
        else:
            print("Word:", node)
getNodes(trees)
```

```python
# S[SEM=(?ad + ?np + WHERE + ?vp)] -> AD[SEM=?ad] NP[SEM=?np] VP[SEM=?vp]
# VP[SEM=(?v + ?pp)] -> IV[SEM=?v] PP[SEM=?pp]
# VP[SEM=(?v + ?ap)] -> IV[SEM=?v] AP[SEM=?ap]
# NP[SEM=(?det + ?n)] -> Det[SEM=?det] N[SEM=?n]
# PP[SEM=(?p + ?np)] -> P[SEM=?p] NP[SEM=?np]
# AP[SEM=?pp] -> A[SEM=?a] PP[SEM=?pp]
# NP[SEM='Country="greece"'] -> 'Greece'
# NP[SEM='Country="china"'] -> 'China'
# Det[SEM='SELECT'] -> 'Which' | 'What'
# N[SEM='City FROM city_table'] -> 'cities'
# IV[SEM=''] -> 'are'
# A[SEM=''] -> 'located'
# P[SEM=''] -> 'in'
hotel_query = 'What are the names of the 5 star hotels in Greece'
hotelsqlgrammer = """
S[SEM=(?np + WHERE + ?vp)] -> AD[SEM=?ad] NP[SEM=?np] VP[SEM=?vp]
VP[SEM=''] -> PP[SEM=?pp] PP[SEM=?ppii]
PP[SEM=(?p + ?np)] -> P[SEM=?p] NP[SEM=?np]
NP[SEM=(?det + ?nn] -> DET[SEM=?det] NUM[SEM=?nu] N[SEM=?nn] 
NP[SEM=''] -> AV[SEM=?av] N[SEM=?n]
NP[SEM=''] -> DET[SEM=?det] N[SEM=?nn]
DET[SEM=''] -> 'the'
N[SEM='hotels FROM hotels_table'] -> 'hotels'
N[SEM=''] ->  'names'
AV[SEM=''] -> 'star'
AD[SEM=(?det + ?iv)] -> DET[SEM=?det] IV[SEM=?iv]
DET[SEM='SELECT'] -> 'What'
IV[SEM=''] -> 'are'
NUM[SEM='5'] -> '5'
NUM[SEM='4'] -> '4'
NUM[SEM='3'] -> '3'
NUM[SEM='2'] -> '2'
NUM[SEM='1'] -> '1'
P[SEM=''] -> 'in' | 'of'
N[SEM='Country="greece"'] -> 'Greece'
"""
with open('sql.fcfg', 'w') as fp:
    fp.write(hotelsqlgrammer)
```
