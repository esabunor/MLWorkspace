```python
def dialogue_act_features(posts, i):
    featuresets = {}
    post = posts[i]
    tokens = nltk.word_tokenize(post.text)
    freq = nltk.FreqDist([ tag for (word, tag) in nltk.pos_tag(tokens)])
    featuresets.update({key: freq[key] for key in freq.keys()})
    for word in tokens:
        featuresets['contains(%s)' % word.lower()] = True
    return [featuresets, post.get('class')]
        
```

```python
size_ = 1000
feature_sets = [dialogue_act_features(nltk.corpus.nps_chat.xml_posts()[:size_], i) for i in range(len(nltk.corpus.nps_chat.xml_posts()[:size_]))]
```

```python
feature_sets
```

```python
len(nltk.corpus.nps_chat.xml_posts())
```

```python
size = int(len(feature_sets) * 0.1)
train_set, test_set = feature_sets[size:], feature_sets[:size]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, test_set))
```

```python
def dialogue_act_features(posts, i):
    featuresets = {}
    post = posts[i]
    tokens = nltk.word_tokenize(post.text)
    freq = nltk.FreqDist([ nltk.map_tag('en-ptb', 'universal', tag) for (word, tag) in nltk.pos_tag(tokens)])
    featuresets.update({key: freq[key] for key in freq.keys()})
    for word in tokens:
        featuresets['contains(%s)' % word.lower()] = True
    return [featuresets, post.get('class')]
```

```python
feature_sets = [dialogue_act_features(nltk.corpus.nps_chat.xml_posts(), i) for i in range(2)]
```
