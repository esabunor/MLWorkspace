```python
def dialogue_act_features(posts, i):
    featuresets = {}
    post = posts[i]
    tokens = nltk.word_tokenize(post.text)
    freq = nltk.FreqDist([ tag for (word, tag) in nltk.pos_tag(tokens)])
    featuresets.update({key: freq[key] for key in freq.keys()})
    for word in tokens:
        featuresets['contains(%s)' % word.lower()] = True
    return [featuresets, post.get('class')]
        
```

```python
dialogue_act_features(nltk.corpus.nps_chat.xml_posts(), 0)
```

```python
feature_sets = [dialogue_act_features(nltk.corpus.nps_chat.xml_posts(), i) for i in range(len(nltk.corpus.nps_chat.xml_posts()))]
```

```python
size = int(len(feature_sets) * 0.1)
train_set, test_set = feature_sets[size:], feature_sets[:size]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, test_set))
```

```python
def rte_features(rtepair):
    extractor = nltk.RTEFeatureExtractor(rtepair)
    features = {}
    features['word_overlap'] = len(extractor.overlap('word'))
    features['word_hyp_extra'] = len(extractor.hyp_extra('word'))
    features['ne_overlap'] = len(extractor.overlap('ne'))
    features['ne_hyp_extra'] = len(extractor.hyp_extra('ne'))
    return features
```

```python
import math
def entropy(labels):
 freqdist = nltk.FreqDist(labels)
 probs = [freqdist.freq(l) for l in nltk.FreqDist(labels)]
 return sum([p * math.log(p,2) for p in probs])
```
